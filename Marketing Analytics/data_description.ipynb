### Simulating Data

##### Store Data: Setting the Structure

```python
import pandas as pd
import numpy as np
```

```python
N_STORES = 40
N_WEEKS = 208
```

```python
columns = ("store_num", "year", "week", "p1_sales", "p2_sales",
           "p1_price", "p2_price", "p1_promo", "p2_promo", "country")
n_rows = N_STORES * N_WEEKS
store_sales = pd.DataFrame(np.empty(shape=(n_rows, 10)), columns=columns)
```

```python
store_sales.shape
```

```python
store_sales.head()
```

```python
store_numbers = range(101, 101 + N_STORES)
list(store_numbers)
```

```python
store_country = dict(zip(store_numbers, ["USA", "USA", "USA", "RSA", "RSA", "RSA",
                                         "RSA", "RSA", "GBR", "GBR", "GBR", "BRA",
                                         "BRA", "JPN", "JPN", "JPN", "JPN", "AUS",
                                         "CHN", "CHN", "RUS", "RUS", "RUS", "BOT",
                                         "BOT", "IND", "IND", "ZIM", "ZIM", "ZIM",
                                         "FRA", "FRA", "FRA", "FRA", "FRA", "FRA",
                                         "USA", "USA", "USA", "RSA"]))
store_country
```

```python
i = 0
for store_num in store_numbers:
    for year in [1, 2, 3, 4]:
        for week in range(1, 53):
            store_sales.loc[i, 'store_num'] = store_num
            store_sales.loc[i, 'year'] = year
            store_sales.loc[i, 'week'] = week
            store_sales.loc[i, 'country'] = store_country[store_num]
            i += 1
```

##### Store Data: Simulating Data Points

```python
np.random.seed(0)
```

```python
store_sales.p1_promo = np.random.binomial(n=1, p=0.1, size=n_rows)
```

```python
store_sales.p2_promo = np.random.binomial(n=1, p=0.15, size=n_rows)
```

```python
store_sales.head(10)
```

```python
store_sales.p1_promo.value_counts()
```

```python
store_sales.p1_price = np.random.choice([30.19, 40.29, 50.49, 60.79, 70.99], size=n_rows)
store_sales.p2_price = np.random.choice([10.29, 20.49, 30.59, 40.99, 50.19], size=n_rows)
store_sales.sample(5)
```

```python
sales_p1 = np.random.poisson(lam=1200, size=n_rows)
sales_p2 = np.random.poisson(lam=1000, size=n_rows)
```

```python
# scale sales according to the ratio of log(price)
log_p1_price = np.log(store_sales.p1_price)
log_p2_price = np.log(store_sales.p2_price)
sales_p1 = sales_p1 * log_p2_price/log_p1_price
sales_p2 = sales_p2 * log_p1_price/log_p2_price
```

```python
# final sales get a 30% or 40% lift when promoted
store_sales.p1_sales = np.floor(sales_p1 * (1 + store_sales.p1_promo * 0.3))
store_sales.p2_sales = np.floor(sales_p2 * (1 + store_sales.p2_promo * 0.4))
store_sales.sample(10)
```

on a series (i.e. a column):

##### Cumulative Distribution

```python
from statsmodels.distributions.empirical_distribution import ECDF
e = ECDF(store_sales.p1_sales)
plt.subplot(1,1,1)
plt.plot(e.x, e.y)
plt.title('Cumulative distribution of p1 weekly sales')
plt.ylabel('Cumulative proportion')
plt.plot([60, 270], [0.9, 0.9], 'k--', alpha=0.5)
plt.plot([store_sales.p1_sales.quantile(.9),
          store_sales.p1_sales.quantile(.9)],
         [0, 1], 'k--', alpha=0.5)
plt.box(False)
```

```python
ecdf_x = store_sales.p1_sales.sort_values()
ecdf_y = np.arange(0, 1, 1/len(store_sales.p1_sales))
plt.subplot(1,1,1)
plt.plot(ecdf_x, ecdf_y)
plt.xlabel('P1 weekly sales, all stores')
plt.ylabel('Cumulative proportion')
plt.plot([60, 270], [0.9, 0.9], 'k--', alpha=0.5)
plt.plot([store_sales.p1_sales.quantile(.9),
          store_sales.p1_sales.quantile(.9)],
         [0, 1], 'k--', alpha=0.5)
plt.box(False)
```
